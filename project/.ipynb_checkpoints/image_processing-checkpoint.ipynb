# Agricultural Pest Identification Using Machine Learning

## Research Questions

1)Can machine learning models accurately classify agricultural pests from images?$

This question explores the feasibility of using ML for pest identification, focusing
on model accuracy and reliability.

2)How do environmental factors influence pest classification accuracy?

Investigates whether attributes like soil type or plant species affect model
performance, aiming to enhance real-world applicability.

3)What are the distinctive features (e.g., size, color) of pests that aid in
classification?

Examines which pest characteristics are most informative for ML models,
providing insights for farmers on key identification traits.

## Motivation

Coming from a country where agriculture is a major occupation, I have seen how pest
infestations cause major crop losses and financial hardship for farmers. Many still rely
on traditional identification methods or broad pesticides, which are often ineffective and
harmful to the environment.
My project aims to address this issue by using machine learning to identify agricultural
pests from images. It will test how accurately models/algorithms can classify pests,
study how environmental factors such as soil type or plant species affect classification,
and examine which visual features like size, color, and shape are most useful for
identification. The backgrounds in the images can also be studied to provide insights
into where certain pests are commonly found, helping connect pest type with habitat
conditions.
By answering these questions, my project aims to create reliable pest detection tools
that support farmers in early diagnosis and better pest management, ultimately
improving crop yield, reducing pesticide misuse,promoting sustainable farming and
reducing food waste.

Rows and columns the dataset

```python
data_dir = Path('data')
extensions = ("*.jpg", "*.jpeg", "*.png")

rows = []
for class_dir in data_dir.iterdir():
    if class_dir.is_dir():
        count = 0
        for ext in extensions:
            count += len(list(class_dir.glob(ext)))
        rows.append({"Class": class_dir.name, "Image Count": count})

df_counts = pd.DataFrame(rows).sort_values("Class").reset_index(drop=True)
print(df_counts)
print("\nTotal images:", df_counts["Image Count"].sum())
```

Plotting the distribution of pests in each class

#### Checking for corrupted files in dataset folder

```python


# Path to your dataset folder
data_dir = Path("data")

# Initialize counters
class_counts = {}
corrupted_files = []

# Loop over all class subfolders (e.g., ants/, bees/, ...)
for class_folder in data_dir.iterdir():
    if class_folder.is_dir():
        image_files = list(class_folder.glob("*.jpg"))
        class_counts[class_folder.name] = len(image_files)
        
        # Check for corrupted or unreadable images
        for img_path in image_files:
            try:
                with Image.open(img_path) as img:
                    img.verify()  # Verifies file integrity without loading full image
            except Exception:
                corrupted_files.append(img_path)

# --- Summary Report ---
print(" Dataset Summary:")
print(f"Total classes: {len(class_counts)}")
print(f"Total images: {sum(class_counts.values())}")
print(f"Corrupted images: {len(corrupted_files)}")

# Create a DataFrame for a nicer table
df_counts = pd.DataFrame(list(class_counts.items()), columns=["Class", "Image Count"])
print("\nImages per Class:")
print(df_counts.to_string(index=False))

# --- Optional: show corrupted file paths ---
if corrupted_files:
    print("\n Corrupted or unreadable image files:")
    for path in corrupted_files:
        print(f" - {path}")
else:
    print("\n No corrupted images found!")
```

Checking for missing values in keras dataset

```python
missing_images = False
missing_labels = False

for batch, labels in data:   # check all batches
    if np.isnan(batch.numpy()).any():
        missing_images = True
    if np.isnan(labels.numpy()).any():
        missing_labels = True

print("Missing image pixel data:", missing_images)
print("Missing label values:", missing_labels)
```

## Visualisation of Data

```python


plt.figure(figsize=(10,6))
plt.bar(df_counts["Class"], df_counts["Image Count"], color='teal')
plt.xticks(rotation=45)
plt.title("Number of Images per Pest Class", fontsize=14)
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.show()
```

```python
import numpy as np
import pandas as pd

# Prepare lists to hold RGB sums and counts per class
class_sums = {c: np.array([0.0, 0.0, 0.0]) for c in data.class_names}
class_counts = {c: 0 for c in data.class_names}

# Iterate over dataset
for images, labels in data:
    for img, label in zip(images.numpy(), labels.numpy()):
        class_name = data.class_names[np.argmax(label)]
        class_sums[class_name] += img.mean(axis=(0,1))  # mean over height & width
        class_counts[class_name] += 1

# Compute averages
avg_rgb_per_class = {c: class_sums[c]/class_counts[c] for c in data.class_names}

# Convert to DataFrame for plotting
df_avg_rgb = pd.DataFrame(avg_rgb_per_class).T  # transpose: classes as rows
df_avg_rgb.reset_index(inplace=True)
df_avg_rgb.rename(columns={'index':'Class', 0:'Red', 1:'Green', 2:'Blue'}, inplace=True)
df_avg_rgb[['Red','Green','Blue']] = df_avg_rgb[['Red','Green','Blue']].astype(float)

print(df_avg_rgb)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

bar_width = 0.2
x = np.arange(len(df_avg_rgb['Class']))

plt.bar(x - bar_width, df_avg_rgb['Red'], width=bar_width, color='red', label='Red')
plt.bar(x, df_avg_rgb['Green'], width=bar_width, color='green', label='Green')
plt.bar(x + bar_width, df_avg_rgb['Blue'], width=bar_width, color='blue', label='Blue')

plt.xticks(x, df_avg_rgb['Class'], rotation=45)
plt.xlabel("Pest Class")
plt.ylabel("Average RGB Value")
plt.title("Average RGB Values per Pest Class")
plt.legend()
plt.show()
```

```python
all_images = []
for images, _ in data.take(5):  # take first 5 batches
    all_images.append(images.numpy())
all_images = np.concatenate(all_images, axis=0)

# Compute mean per channel for each image
mean_per_channel = all_images.mean(axis=(1,2))
df_channels = pd.DataFrame(mean_per_channel, columns=['Red','Green','Blue'])

# Correlation heatmap
plt.figure(figsize=(6,5))
sns.heatmap(df_channels.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between RGB Channels")
plt.show()
```
